```python
data_1 = pd.read_csv('beta_overall_1901-1950_20clus_50years_50op_rs.dat',
                     sep='\s+',header=None)
data_2 = pd.read_csv('overall_locations_analysis/beta_overall_1901-195050years_intesection_ovverall.dat',
                     sep='\s+',header=None)
```

```python
beta_op = data_1.iloc[0][0:-1].values
beta_ov = data_2.iloc[0][0:-1].values
```

```python
beta_ov.shape
```

```python
loc_1 = pd.read_csv('output_1901-1950_50years_20clus_50op_rs.dat',sep='\s+',header=None).values
loc_2 = pd.read_csv('intersection_overall.dat',sep='\s+',header=None).values
```

```python
loc1 = np.asarray(loc_1[:,0:-7].reshape(1,-1),dtype='int32') - 1
loc2 = np.asarray(loc_2.reshape(1,-1),dtype='int32') - 1
```

```python
loc2.shape
```

```python
lon_list_1 = total_latlist[loc1].reshape(1000)
lat_list_1 = total_lonlist[loc1].reshape(1000)

lon_list_2 = total_latlist[loc2].reshape(1005)
lat_list_2 = total_lonlist[loc2].reshape(1005)
```

```python
lon_list_1.shape
```

```python
dr1 = xr.DataArray(beta_op,dims=("op"),coords={"op":loc1.reshape(-1)})
```

```python
dr1.coords['lat'] = ('op',lat_list_1)
dr1.coords['lon'] = ('op',lon_list_1)
```
